#### Stage 1: Dividing the population into 4 strata (Stratified Sampling Stage 1)

Pembagian ini dilakukan karena asumsi awal bahwa terdapat perbedaan pengangguran berdasarkan jenjang pendidikan. Maka penduduk dibagi menjadi SMP/SD Ke bawah/ DIV,S1,S2,S3/DI,II,III/SMK/SMA. Persentase rata-rata diperoleh dari data BPS DKI Jakarta.

Total Cost Formula :

$$
C = c_{0} + c_{1}n + c_{2}nm
$$

```python
# Assumption Formula
c0 = 4_000_000
c1 = 50_000
c2 = 50_000
N = 5
n_ = 5
M = 1000
nm = 1000
```


$$
m_{\text{opt}} = \sqrt{
\cfrac{c_{1} \sigma_{w}^{2}}
{c_{2} \left( \sigma_{b}^{2} - \cfrac{\sigma_{w}^{2}}{\bar{M}} \right)}
}
$$

$$
\sigma_{b}^{2} = \cfrac{1}{N-1} \sum_{i=1}^{N} (\mu_{i} - \mu)^{2}
$$

$$
\sigma_{w}^{2} = \cfrac{1}{N} \sum_{i=1}^{N} \sigma_{i}^{2}
$$

```python
# Average Revenue
rev_dki_2021 = rev_kota_2021.loc[6::,["Rata-rata"]]
rev_dki_2021 = rev_dki_2021["Rata-rata"].values[0]
```

```python
rev_kota_2021["Variance"] = (rev_kota_2021["Rata-rata"]-rev_dki_2021)**2
```

```python
rev_kota_2021
```

Kalkulasi Performaed untuk mencari find σ_b^2 dan σ_w^2

Karena kita sudah menghitung varians data, kita sekarang dapat menghitung σ_b^2 dan σ_w^2 berdasarkan rumus di atas. Akhirnya, kita bisa mendapatkan nilai untuk m.opt

```python
# Value of sigma_b^2 & sigma_w^2
sigma_b = (1/(N-1))*(rev_kota_2021["Variance"].sum())
sigma_w = (1/(N))*(rev_kota_2021["Variance"].sum())
print("Nilai dai sigma_b^2 adalah", sigma_b)
print("Nilai dai sigma_w^2 adalah", sigma_w)
```

```python
# Calculate value of m_opt
m_opt = math.sqrt(c1*sigma_w/(c2*(sigma_b-(sigma_w/M))))
print("The value of m_opt is", m_opt)
```

Langkah selanjutnya adalah menghitung total biaya alokasi (C) dengan 5 PSU, yang sesuai dengan jumlah kota yang dipilih dan mengasumsikan ukuran sampel data sebelumnya sebesar 1000 (data yang diasumsikan). Hal ini dapat dihitung sebagai berikut:

```python
# Calculate total allocation cost with C formula
C = c0 + c1*n_+c2*nm
print("Total allocation cost is",C)
```


Total biaya alokasi telah dihitung, jadi kita dapat menggunakan persamaan berikut untuk menentukan ukuran sampel yang diperlukan.

$$
n = \cfrac{C - c_{0}}{c_{1} + c_{2} m_{\text{opt}}}
$$

```python
# Calculate total sample or n
n = (C-c0)/(c1+c2*m_opt)
n = math.ceil(n)
print("Total sample needed is", n)
```

### F.  Estimating the Needs of the Number of Samples

Metode yang digunakan untuk pengumpulan data adalah melalui telepon atau menggunakan layanan pesan daring. Metode pengumpulan data dengan sampling telepon merupakan salah satu cara untuk mengumpulkan data melalui telepon. Metode ini merupakan metode yang populer dalam melakukan survei dan riset pasar, karena relatif lebih murah dan cepat dibandingkan dengan metode lain seperti wawancara langsung atau penyebaran kuesioner.

Untuk menggunakan metode ini, Anda harus terlebih dahulu membuat daftar sampel yang akan dihubungi. Sampel ini dapat diambil secara acak dari populasi yang akan diteliti. Setelah itu, surveyor atau pewawancara akan menghubungi setiap nomor telepon yang ada dalam sampel dan mengajukan serangkaian pertanyaan yang diperlukan untuk mengumpulkan data.

Beberapa hal yang perlu diperhatikan dalam metode ini adalah:

1. Validitas Sampel: Pastikan bahwa sampel yang diambil benar-benar mewakili populasi yang ingin diteliti.
2. Keterampilan Pewawancara: Pewawancara harus memiliki keterampilan yang baik dalam mengelola wawancara telepon dan memastikan bahwa jawaban dari responden valid dan akurat.
3. Biaya: Meskipun relatif lebih murah dibandingkan dengan metode lain, biaya tetap dapat menjadi faktor yang signifikan tergantung pada jumlah sampel yang ingin diteliti.
4. Responden Tidak Tersedia: Beberapa responden mungkin tidak tersedia atau tidak bersedia berpartisipasi dalam wawancara telepon.
5. Keabsahan Jawaban: Jawaban yang diberikan oleh responden mungkin tidak akurat karena mereka mungkin tidak ingin terbuka atau merasa tidak nyaman dengan wawancara telepon.
